### Truy vấn lọc bỏ ra và xóa đi những dòng có cầu thủ bị ít thẻ đỏ, thẻ vàng (< 80%)

+ Những cầu thủ có số lần bị thẻ vàng, thẻ đỏ < 80% sẽ được loại bỏ và xóa
+ Việc làm như vậy sẽ giúp dữ liệu dễ dàng trong việc tìm kiếm 

```python
for i in range (len(Question_4)):
    if Question_4['Normalize_YelCard'][i] <= 0.8 and Question_4['Normalize_RedCard'][i] <= 0.8  :
        Question_4=Question_4.drop(i, axis =0)
```

```python
Question_4
```

```python
Question_4.mean()
```

### Những bước tiếp theo:
- Lọc lại danh sách cầu thủ có số thẻ vàng lớn >= 2 hoặc là có số thẻ đỏ > 0 


```python
Not_FairPlay= Question_4.loc[ (Question_4.YelCard >= 2) | (Question_4.RedCard > 0) ,['Player','YelCard','RedCard']]
```

```python
Not_FairPlay
```

- Theo ta được biết thì 2 thẻ vàng trong một trận đấu tương đương với 1 thẻ đỏ. Nhưng trong tập dữ liệu chỉ là thống kê tổng cộng thẻ vàng, thẻ đỏ.
- Vì vậy giả định 2 thẻ vàng bằng 1 thẻ đỏ. Sau đó tính lại tổng số thẻ đỏ lưu vào Sum_Yel_Red_Card 


```python
# Giả định: 2 thẻ vàng = 1 thẻ đỏ.
Not_FairPlay['Sum_Yel_Red_Card'] = Not_FairPlay['YelCard']/2 + Not_FairPlay['RedCard']
```

- Sau khi đã lưu lại thì tiếp tục lọc ra những cầu thủ có số thẻ đỏ lớn bằng 3 ( dựa vào Sum_Yel_Red_Card )

```python
temp= Not_FairPlay.loc[Not_FairPlay.Sum_Yel_Red_Card >= 3]
temp
```

```python
plt.figure(figsize=(15, 7))


plt.barh(Not_FairPlay['Player'],Not_FairPlay['Sum_Yel_Red_Card'],color='#33CC99')
plt.ylabel('Player')
plt.xlabel("Sum Card (Yellow, Red)")
plt.title("Some Players Not FairPlay");
```

### Để tìm ra đội hình phù hợp nhất thì nhóm đã lựa chọn những tiêu chí sau:
- Tiền đạo phải ghi được nhiều bàn thắng, hạn chế bị phạm lỗi. 
- Tiền vệ và hậu vệ phải hỗ trợ nhiều và ít lỗi nhất (thẻ đỏ, thẻ vàng).
- Thủ môn bắt bóng tốt và không có thẻ đỏ và thẻ vàng .

### Chuẩn hóa dữ liệu : Goals, Appears, Assitst

```python
Question_6['Normalize_Goals']=(Question_6['Goals']- Question_6['Goals'].min()) / (Question_6['Goals'].max() - Question_6['Goals'].min())

Question_6['Normalize_Appears']=(Question_6['Apps']- Question_6['Apps'].min()) / (Question_6['Apps'].max() - Question_6['Apps'].min())

Question_6['Normalize_Assists']=(Question_6['Assists']- Question_6['Assists'].min()) / (Question_6['Assists'].max() - Question_6['Assists'].min())

Question_6
```

- Dể dữ liệu Position như vậy khó phân tích vị trí của các cầu thủ trên sân.

- Sau đó xóa đi cột Position

```python
Question_6['Position'] = Question_6['Position'].apply(lambda genres: genres.split(","))
from sklearn.preprocessing import MultiLabelBinarizer


mlb = MultiLabelBinarizer()
Question_6 = Question_6.join(pd.DataFrame(mlb.fit_transform(Question_6.pop('Position')),
                          columns=mlb.classes_,
                          index=Question_6.index))

Question_6
```
